```python
import requests as rq
from bs4 import BeautifulSoup as bs
import pandas as pd
import numpy as np                                        
import matplotlib.pyplot as plt
```

```python
headers = ({'User-Agent' : 'Mozilla/5.0 (X11; Ubuntu; Linux x86_64; rv:102.0) Gecko/20100101 Firefox/102.0'})
url ='https://www.olx.com.br/imoveis/aluguel/estado-df'
response = rq.get(url,  headers=headers)
content = response.content

site = bs(content, 'html.parser')
```

```python
todas_paginas = site.find('div', class_='sc-EHOje gpDieb')

paginas = []


for item in todas_paginas.find_all('a'):
    paginas.append(item['href'])
paginas
```

```python
links = []


itens = site.find('ul', id='ad-list')

for pagina in paginas:
    for item in itens.find_all('a'):
        links.append(item['href'])
    else:
        links.append('') 
```

```python
len(links)
```

```python
valor_aluguel = []


itens = site.find('ul', id='ad-list')
for pagina in paginas:
    for item in itens.find_all('span', class_='main-price sc-ifAKCX bytyxL'):
            valor_aluguel.append(item['aria-label'])
    else:
        valor_aluguel.append('') 
```

```python
boxes = site.find('ul', id='ad-list')
```

```python
titulos = []


itens = site.find('ul', id='ad-list')
for pagina in paginas:
    for item in itens.find_all('h2'):
        titulos.append(item['aria-label'])
    else:
        titulos.append('') 
```

```python
len(titulos)
```

```python
list_qtde = []
for pagina in paginas:
    for box in boxes:
        list_aux = []
        for item in box.find_all('li'):

            for item_ in item:
                if item_.get('aria-label'):
                    list_aux.append(item_.get('aria-label'))
                else:
                     list_aux.append('')

        list_qtde.append(list_aux)
```

```python
lista = []

itens = site.find('ul', id='ad-list')
for pagina in paginas:
    for item in itens.find_all('div', class_='sc-iNhVCk geaanK'):
        lista.append(item.text)
    else:
        lista.append('') 
```

```python
len(lista)
```

```python
df = pd.DataFrame(list(zip(titulos, valor_aluguel, links, list_qtde, lista)),
                  columns=['titulo', 'preço', 'links', 'list', 'lista'])
```

```python
def quartos(row):
    """pegar quantidade de quatos da lista e criar nova coluna

    Args:
        row (list): Lista que contem todas as informacoes

    Returns:
        list: list com a quantidade de quartos
    """
    row = list(filter(lambda x: 'quarto' in x, row))
    return row


def metros_quadrado(row):
    """pegar metros quadrados da lista e criar nova coluna

    Args:
        row (list): Lista que contem todas as informacoes

    Returns:
        list: list com a quantidade de metros quadrados
    """
    row = list(filter(lambda x: 'metros quadrados' in x, row))
    return row

def banheiro(row):
    """pegar metros quadrados da lista e criar nova coluna

    Args:
        row (list): Lista que contem todas as informacoes

    Returns:
        list: list com a quantidade de metros quadrados
    """
    row = list(filter(lambda x: 'banheiro' in x, row))
    return row
```

```python
df['quartos'] = df['list'].apply(quartos)
df['metro_quadrado'] = df['list'].apply(metros_quadrado)
df['banheiro'] = df['list'].apply(banheiro)
```
